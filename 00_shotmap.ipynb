# Shot Map for 2018 World Cup Final (France v. Croatia)

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager 
from mplsoccer import Pitch
from mplsoccer.cm import create_transparent_cmap
from mplsoccer.scatterutils import arrowhead_marker
from statsbombpy import sb
```

```python
# See the choice of competition sets available
sb.competitions()
```

```python
# Get the 2018 World Cup set
sb.matches(competition_id=43, season_id=3)
```

```python
# Get matches
wc_2018_subset = sb.matches(competition_id=43, season_id=3)[["match_id","match_date","home_team", "away_team", "home_score", "away_score"]]
```

```python
wc_2018_subset
```

```python
# Get all the France games
france_games = wc_2018_subset[(wc_2018_subset["home_team"] == 'France') | (wc_2018_subset["away_team"] == 'France')] 
```

```python
# Put the games in order by date
france_games.sort_values('match_date')
```

```python
# get the last game for France 
events_8658 = sb.events(match_id=8658)
```

```python
# Look at one of the team lineups for the game to determine which columns you need to mark shots
sb.lineups(match_id=8658)["France"].columns
```

```python
# Create lineup DataFrames with player ID and Jersey Number
france_lineup = sb.lineups(match_id=8658)['France'][['player_id', 'jersey_number']]
croatia_lineup = sb.lineups(match_id=8658)['Croatia'][['player_id', 'jersey_number']]
```

```python
france_lineup
```

```python
# look at the event data columns to see the shot data you need
events_8658.info()
```

```python
# Get the event columns you need
events_8658_sub = events_8658[['type',"team", "player_id", "shot_outcome",'shot_statsbomb_xg','shot_type','location','minute', 'second', 'timestamp']]
```

```python
france_shots = events_8658_sub[(events_8658_sub['type'] == 'Shot') & (events_8658_sub['team'] == "France")].sort_values("minute")
```

```python
# At the end of the game 3009 put a ball right at 2004's feet, right in front of the goal
#    and he missed the ball with his foot. Shouldn't that be added to xG?
# or is that added to xT?
france_shots
```

```python
france_lineup
```

```python
# add jersey_number column from f_player_info to the france_shots DF
# this is a default inner join
france_shots = pd.merge(france_shots, france_lineup)
```

```python
france_shots.sort_values("minute")
```

```python
# Seperate x and y location values into different columns
france_shots[['x','y']] = (france_shots['location'].tolist())
```

```python
france_shots
```

```python
# move the jersey_number column next to the player_id column
col = france_shots.pop('jersey_number')
france_shots.insert(3, 'jersey_number', col)
```

```python
france_shots
```

```python
# Do the same for Croatia
croatia_shots = events_8658_sub[(events_8658_sub['type'] == 'Shot') & (events_8658_sub['team'] == "Croatia")].sort_values("minute")
```

```python
croatia_shots = pd.merge(croatia_shots, croatia_lineup)
```

```python
croatia_shots[['x','y']] = (croatia_shots['location'].tolist())
```

```python
col = croatia_shots.pop('jersey_number')
croatia_shots.insert(3, 'jersey_number', col)
```

# Create Shot Maps

```python
fpitch = Pitch(pitch_color='grass', line_color='white', stripe=True)

# Make the diffrent shot info 

france_shots_goals = france_shots[france_shots['shot_outcome'] == 'Goal']
france_shots_on = france_shots[(france_shots['shot_outcome'] != 'Goal') & (france_shots['shot_outcome'] != 'Off T')]
france_shots_off = france_shots[france_shots['shot_outcome'] == 'Off T']


fig, ax = fpitch.draw(figsize=(12, 10))
#sc1 = fpitch.scatter(france_shots_goals.x, france_shots_goals.y, s=500, c='white', marker = 'football', alpha = .9, ax=ax)
sc1 = fpitch.scatter(france_shots_goals.x, france_shots_goals.y, s=500, c='white', alpha = 1, ax=ax)
sc2 = fpitch.scatter(france_shots_on.x, france_shots_on.y, s=500, c='lightblue', alpha = 1, ax=ax)
sc3 = fpitch.scatter(france_shots_off.x, france_shots_off.y, s=500, c='orange', alpha = 1, ax=ax)

# plot the jersey numbers to the shots
for i, label in enumerate(france_shots.jersey_number):
    sc4 = fpitch.annotate(label, (france_shots.x[i], france_shots.y[i]),
    va='center', ha='center', color='black',
    fontproperties ='comic.ttf', fontsize=15, ax=ax)
    
txt = ax.text(x=40, y=80, s='France shots v. Croatia\n white=goal\n blue=on target\n yellow= off target',
             size=20, color='black', weight='bold',
              va='bottom', ha='right')

# make a legend to shoiw shot colors
# mark saved shots
# mark blocked shots 
# mark wayward shots
# mark own goals red on the opposition's shot chart (Croatia's on France's)
# I want to put a link on each shot to the video clip (for context)

# No Corners and Free kick info - which resulted in the first three goals
```

```python
cpitch = Pitch(pitch_color='grass', line_color='white', stripe=True)

# Make the diffrent shot info 

croatia_shots_goals = croatia_shots[croatia_shots['shot_outcome'] == 'Goal']
croatia_shots_on = croatia_shots[(croatia_shots['shot_outcome'] != 'Goal') & (croatia_shots['shot_outcome'] != 'Off T')]
croatia_shots_off = croatia_shots[croatia_shots['shot_outcome'] == 'Off T']

fig, ax = cpitch.draw(figsize=(12, 10))
#sc1 = cpitch.scatter(croatia_shots_goals.x, croatia_shots_goals.y, s=500, c = 'white', marker = 'football', ax=ax)
sc1 = cpitch.scatter(croatia_shots_goals.x, croatia_shots_goals.y, s=500, c = 'white', alpha = 1, ax=ax)
sc2 = cpitch.scatter(croatia_shots_on.x, croatia_shots_on.y, s=500, c='lightblue', alpha = 1, ax=ax)
sc3 = cpitch.scatter(croatia_shots_off.x, croatia_shots_off.y, s=500, c='yellow', label = '10', alpha = 1, ax=ax)
# plot the jersey numbers to shots
for i, label in enumerate(croatia_shots.jersey_number):
    sc4 = cpitch.annotate(label, (croatia_shots.x[i], croatia_shots.y[i]),
    va='center', ha='center', color='black',
    fontproperties ='comic.ttf', fontsize=15, ax=ax)  
    
text = ax.text(x=40, y=80, s='Croatia shots v. France\n white=goal\n blue=on target\n yellow= off target',
             size=20, color='black', weight='bold',
             va='bottom', ha='right')

```
